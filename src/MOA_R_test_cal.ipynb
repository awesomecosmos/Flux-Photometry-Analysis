```python
import calibrimbore as cal 

import numpy as np
import matplotlib.pyplot as plt
from glob import glob
%matplotlib inline

from astropy.coordinates import SkyCoord
import astropy.units as u
from copy import deepcopy
```

```python
R_filter = 'moared.txt'
R_fit = cal.sauron(band=R_filter,system='skymapper',gr_lims=[-.5,0.8],plot=True,cubic_corr=False)
```

```python
# coords = ['08:05:11.1 -11:32:01.9']
coords = ['05:54:26.71 -56:01:38.3']  
c = SkyCoord(coords,unit=(u.hourangle, u.deg))
sources = cal.get_skymapper_region(c.ra.deg,c.dec.deg)

ind = (np.isfinite(sources.r.values) & np.isfinite(sources.i.values) 
       & np.isfinite(sources.z.values) & (sources['g'].values < 19) & (sources['g'].values > 13))
sources = sources.iloc[ind]
```

```python
sources
```

```python
R_estimates = R_fit.estimate_mag(mags = sources)

sources['MOA_R_est'] = R_estimates
```

```python
sources
```

```python
plt.figure()
plt.plot(sources['g']-sources['r'],sources['MOA_R_est']-sources['r'],'.',color="darkviolet")
plt.xlabel("g-r")
plt.ylabel("MOA-R - r")
plt.grid("both")
plt.title("(g-r) vs (MOA-R - r)")
```
